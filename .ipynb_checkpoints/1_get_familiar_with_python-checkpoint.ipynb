```python
## read text file
#fname = 'FernandinaSenDT128/bl_list.txt'
fname = '/Users/yunjunz/development/python/a_python_guide_to_geophysics/FernandinaSenDT128/bl_list.txt'
f = open(fname, 'r')
lines = f.readlines()
f.close()
```

```python
fcontent = np.loadtxt(fname, dtype=str)
date6_list = fcontent[:, 0]
bperps = fcontent[:, 1].astype(np.float)
```

```python
import matplotlib.pyplot as plt
plt.figure()
plt.plot(bperps)
plt.show()
```

```python
from datetime import datetime as dt
date8_list = ['20{}'.format(i) for i in date6_list]
dates = [dt.strptime(i, '%Y%m%d') for i in date8_list]
plt.figure()
plt.plot(dates, bperps, '-o')
plt.xlabel('Time')
plt.ylabel('Satellite Relative Orbit Separation')
plt.show()
```
