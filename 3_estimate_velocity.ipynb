```python
## Estimate Linear Velocity from 3D time-series displacement on Fernandina Volcano

import os
from datetime import datetime as dt
import h5py
import numpy as np
import matplotlib.pyplot as plt

work_dir = os.path.expanduser('~/development/python/a_python_guide_to_geophysics/FernandinaSenDT128')
os.chdir(work_dir)
print('Go to directory:', work_dir)
```

```python
# read timeseries HDF5 file
ts_file = 'timeseries.h5'
f = h5py.File(ts_file, 'r')
ts_data = f['timeseries'][:]
date_list = [i.decode('utf8') for i in f['date'][:].tolist()]
f.close()
```

```python
## prepare design matrix
dates = [dt.strptime(i, '%Y%m%d') for i in date_list]
years = np.array([i.year + (i.timetuple().tm_yday - 1)/365.25 for i in dates])
A = np.ones((len(years), 2), dtype=np.float32)
A[:, 0] = years
A_inv = np.dot(np.linalg.inv(np.dot(A.T, A)), A.T)
```

```python
num_date, length, width = ts_data.shape
vel = np.dot(A_inv, ts_data.reshape(num_date, -1))[0, :].reshape(length, width)
```

```python
plt.figure()
plt.imshow(vel)
plt.show()
```

```python
y, x = 235, 140
pts_ts = ts_data[:, y, x]
pts_vel = vel[y, x]

off = np.dot(A_inv, ts_data.reshape(num_date, -1))[1, :].reshape(length, width)
pts_ts_lin = (years - years[0]) * pts_vel + off[y, x] / years[0]

plt.figure()
plt.plot(dates, pts_ts, '.')
plt.plot(dates, pts_ts_lin, '--')
plt.show()
```
